#### Punto b

Agarramos toda la tabla de Transfer Market y vamos a las columnas que tiene la infomacion del jugador a nivel personal y de clubes, luego vamos a realizar un label de las posiciones y del pie dominante

```python
data = transferMarket.drop(columns=["Column1","nationality", "squad", "birth_year", "position2", "league", "Season", "Attendance", "CL","WinCL","CLBestScorer"])

Labels = LabelEncoder()

data["position"] = Labels.fit_transform(data["position"])
data["position2"] = Labels.fit_transform(data["position2"])
data["foot"] = Labels.fit_transform(data["foot"])
```

Realizamos el escalado y nos guardamos los scalers para despues preparar las predicciones

```python
scalerX = MinMaxScaler().set_output(transform="pandas")
scalerY = MinMaxScaler().set_output(transform="pandas")



X = data.drop(columns=["player", "value"])
y = data["value"]

# Dividimos amientes de testeo y de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

#Escalamos X y lo Centramos
X_train = scalerX.fit_transform(X_train)
X_test = scalerX.transform(X_test)

train_mean = X_train.mean()
X_train -= train_mean
X_test -= train_mean

#Convertimos a dataframe para escalar
y_train = pd.DataFrame(y_train)
y_test = pd.DataFrame(y_test)

#Escalamos
y_train = scalerY.fit_transform(y_train)
y_test = scalerY.transform(y_test)

#Lo pasamos nuevamente a Serie
y_train = y_train["value"]
y_test = y_test["value"]

X_train
```

Preparamos el modelo para el analisis y le realizamos el fit, ademas vamos a ver los grafico de error

```python
keras.utils.set_random_seed(11)

# Paso 1: iniciamos el modelo indicando la arquitectura de la red
model = keras.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='relu'),
])

# Paso 2: configuramos el optimizador
optimizer = keras.optimizers.SGD(learning_rate=0.2)

# Paso 3: construimos el modelo
model.compile(
    optimizer=optimizer,
    loss='mean_squared_error',
    metrics=["mean_absolute_error"]
)

# Paso 4: entrenamiento del modelo.
hist = model.fit(X_train.to_numpy(), y_train.to_numpy(),
                 epochs=50,
                 batch_size=20,
                 validation_split=0.2,
                 verbose=1,
                 )

# Paso 5: calculamos el error en el conjunto de testeo
model.evaluate(X_test.to_numpy(), y_test.to_numpy(),
               verbose=1,
               batch_size=len(y_test),
               return_dict=True
               )
```

```python
# Armamos una funcion para graficar el error a través de las épocas
def graficar_error(history, error_name):
    x_arr = np.array(history.epoch)    # en el atributo epoch, history guarda una lista de epocas
    plot = (
        so.Plot()
        .add(so.Line(color='blue'), x=x_arr, y=history.history[error_name], label='Entrenamiento')
        .add(so.Line(color='orange'), x=x_arr, y=history.history[f'val_{error_name}'], label='Validacion')
        .label(title=error_name)
    )
    plot.show()

graficar_error(hist, 'loss')
```

Ahora que tenemos nuestro modelo, vamos a escalar todo X y pasarlo por el modelo

```python
pd.options.display.float_format = '{:.0f}'.format

# Escalamos y centramos X a nuestro escalado para calcular para todos
X_scaled = scalerX.transform(X) - train_mean

X_scaled
```

```python
# Predecimos segun nuestro modelo y lo escalamos inversamente a Y
preciosPredecidos = scalerY.inverse_transform(model.predict(X_scaled))

preciosPredecidos
```

Ahora realizamos la tabla de precios y su diferencia
